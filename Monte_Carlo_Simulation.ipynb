```python
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def spinner():
    slot = random.randint(0,36)
    if slot == 0:
        return 'lost'
    elif 1 <= slot <= 18:
        return 'won'
    elif slot > 18:
        return 'lost'
```

```python
def simple_bettor(budget, bet, periods):
    x_axis = []
    y_axis = []
    
    currentPeriod = 1
    while currentPeriod <= periods:
        result = spinner()
        if result == 'won':
            budget = budget + bet
        elif result == 'lost':
            budget = budget - bet
        x_axis.append(currentPeriod)
        y_axis.append(budget)
        
        currentPeriod = currentPeriod + 1
    plt.plot(x_axis, y_axis)
    return y_axis[-1]
```

```python
futures = 1
outcomes = []
while futures < 101:
    outcomes.append(simple_bettor(1000, 100, 1000))
    #the larger the number of the periods the more downward oriented trend will be observed
    futures = futures + 1

plt.ylabel('Budget')
plt.xlabel('Periods')
plt.title('Average loser')
#distribution of possible future budgets, vertical argument is for transposing
```

```python
sns.distplot(outcomes, bins = 20, color = 'green') 
```

```python
def smart_bettor(budget, bet, periods): #smart bettor is doubling his bet in case of losing
    x_axis = []
    y_axis = []
   
    
    currentPeriod = 1
    initial = bet
    while currentPeriod <= periods:
        result = spinner()
        if result == 'won':
            budget = budget + bet
            bet = initial
        elif result == 'lost':
            budget = budget - bet
            bet = 2 * bet
            
        x_axis.append(currentPeriod)
        y_axis.append(budget)
        
        currentPeriod = currentPeriod + 1
    plt.plot(x_axis, y_axis)
    return y_axis[-1]
```

```python
futures = 1
outcomes = []
while futures < 101:
    outcomes.append(smart_bettor(1000, 100, 1000))
    #the larger the number of the periods the more downward oriented trend will be observed
    futures = futures + 1

plt.ylabel('Budget')
plt.xlabel('Periods')
plt.title('Average winner')
```

```python
sns.distplot(outcomes, bins = 20, color = 'green') 
```
